```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plotting style
plt.style.use('seaborn-v0_8-muted')
sns.set_theme(style="whitegrid")

# Load the data
data_path = "../backend/data/raw/sale_clean.csv"
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
else:
    print("Dataset not found. Please ensure backend/data/raw/sale_clean.csv exists.")
    df = pd.DataFrame()
```

### 1. Target Variable Distribution: Real Estate Prices in Italy
We analyze the distribution of `prezzo` (price) to understand its spread and see if a log-transformation might be beneficial during training.


```python
if not df.empty:
    plt.figure(figsize=(12, 6))
    sns.histplot(df['prezzo'], bins=50, kde=True, color='indigo')
    plt.title("Distribution of Property Prices (Euro)", fontsize=15)
    plt.xlabel("Price (€)", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.show()

    # Log Transformation to handle skewness
    plt.figure(figsize=(12, 6))
    sns.histplot(np.log1p(df['prezzo']), bins=50, kde=True, color='teal')
    plt.title("Distribution of Log-Price", fontsize=15)
    plt.xlabel("Log(Price)", fontsize=12)
    plt.show()
else:
    print("No data to plot distribution.")
```

### 2. Feature Correlation: Finding Price Drivers
Which property attributes correlate most strongly with the target price? (Area, Rooms, Bathrooms, etc.)


```python
if not df.empty:
    numeric_df = df.select_dtypes(include=[np.number])
    corr_matrix = numeric_df.corr()
    
    # Top 10 correlates with Price
    top_corr = corr_matrix['prezzo'].sort_values(ascending=False).head(10)
    print("Most Influential Features with Price:")
    print(top_corr)

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=False, cmap='RdYlBu', linewidths=0.5)
    plt.title("Correlation Heatmap: Italy Housing Attributes", fontsize=15)
    plt.show()
else:
    print("No data for correlation analysis.")
```

### 3. Surface Area vs Price: Mapping Property Value per Square Meter
Visualization of Area vs. Price, typically the most powerful predictor in real estate.


```python
if not df.empty and 'superficie' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='superficie', y='prezzo', data=df, alpha=0.5, color='darkorange')
    plt.xscale('log')
    plt.yscale('log')
    plt.title("Log-Log Relationship: Surface Area vs Price", fontsize=15)
    plt.xlabel("Surface Area (m²)", fontsize=12)
    plt.ylabel("Price (€)", fontsize=12)
    plt.show()
else:
    print("Column 'superficie' not available.")
```
